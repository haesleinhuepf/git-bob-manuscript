```python
issue_numbers = list(range(121, 141))
issue_numbers
```

```python
from git_bob._github_utilities import get_most_recent_comment_on_issue
```

```python
get_most_recent_comment_on_issue(repository="haesleinhuepf/stackview", issue=121)
```

```python
answers = []
links = []
contains_haesleinhuepf = []
for i in issue_numbers:
    author, response = get_most_recent_comment_on_issue(repository="haesleinhuepf/stackview", issue=i)
    answers.append(response)
    links.append(f"https://github.com/haesleinhuepf/stackview/issues/{i}")
    contains_haesleinhuepf.append("@haesleinhuepf" in response)
```

```python
import pandas as pd
df = pd.DataFrame({
    "questions":questions,
    "answers":answers,
    "links":links,
    "contains_haesleinhuepf": contains_haesleinhuepf
})
df
```
